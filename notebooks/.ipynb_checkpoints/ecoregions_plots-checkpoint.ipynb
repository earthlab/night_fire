```python
import geopandas as gpd
from glob import glob
import os
from matplotlib import pyplot as plt
from helpers import extract_frp_day_night_annual, customized_box_plot
import pandas as pd
import numpy as np
```

```python
fire_pts_folders = !ls -d ../epa_L1_ecoregion_points/*
root_dir = '../epa_L1_ecoregion_points*'
# dbf_files = [glob(folder + '/*.dbf')[0] for folder in fire_pts_folders]
dbf_files = glob(root_dir + '/*.dbf')
```

```python
frps = []
conf=0
for dbf in dbf_files:
    frps.append(extract_frp_day_night_annual(dbf)) #default conf=50
```

```python
df = pd.DataFrame(frps).sort_values('years')
df
```

```python
def customized_box_plot(percentiles, axes, redraw = True, *args, **kwargs):
    """
    Generates a customized boxplot based on the given percentile values
    """

    n_box=percentiles.shape[0]
    box_plot = axes.boxplot([[-9, -4, 2, 4, 9],]*n_box, *args, **kwargs) 
    # Creates len(percentiles) no of box plots

    min_y, max_y = float('inf'), -float('inf')

    fliers_xy=None
    for box_no, (q1_start, 
                 q2_start,
                 q3_start,
                 q4_start,
                 q4_end) in enumerate(percentiles):

        # Lower cap
        box_plot['caps'][2*box_no].set_ydata([q1_start, q1_start])
        # xdata is determined by the width of the box plot

        # Lower whiskers
        box_plot['whiskers'][2*box_no].set_ydata([q1_start, q2_start])

        # Higher cap
        box_plot['caps'][2*box_no + 1].set_ydata([q4_end, q4_end])

        # Higher whiskers
        box_plot['whiskers'][2*box_no + 1].set_ydata([q4_start, q4_end])

        # Box
        box_plot['boxes'][box_no].set_ydata([q2_start, 
                                             q2_start, 
                                             q4_start,
                                             q4_start,
                                             q2_start])

        # Median
        box_plot['medians'][box_no].set_ydata([q3_start, q3_start])

        # Outliers
#         if fliers_xy is not None and len(fliers_xy[0]) != 0:
#             # If outliers exist
#             box_plot['fliers'][box_no].set(xdata = fliers_xy[0],
#                                            ydata = fliers_xy[1])

#             min_y = min(q1_start, min_y, fliers_xy[1].min())
#             max_y = max(q4_end, max_y, fliers_xy[1].max())

#         else:
#             min_y = min(q1_start, min_y)
#             max_y = max(q4_end, max_y)

        # The y axis is rescaled to fit the new box plot completely with 10% 
        # of the maximum value at both ends
#         min_y = min(q1_start, min_y)
#         max_y = max(q4_end, max_y)

#         axes.set_ylim([min_y*1.5, max_y*1.5])

    min_y = min(percentiles.min(), min_y)
    max_y = max(percentiles.max(), max_y)

    #axes.set_ylim([min_y*0.5, max_y*1.25])
    axes.set_ylim([0, percentiles.max()+20])
    # If redraw is set to true, the canvas is updated.
    if redraw:
        ax.figure.canvas.draw()

    return box_plot
```

```python
## day
# percentiles = np.array([frp_day_5, frp_day_25, frp_day_median, frp_day_75, frp_day_95]).T
percentiles = np.array([df.day_frp_5.values, df.day_frp_25.values, df.day_frp_median.values, df.day_frp_75.values, df.day_frp_95.values]).T
print(percentiles.shape)
year = df.years.values

fig, ax = plt.subplots(figsize=(15,10))
b = customized_box_plot(percentiles, ax, redraw=True, notch=0, sym='+', vert=1, whis=1.5)
ax.set_ylabel('FRP ($W/m^2$)', fontsize=12)
ax.set_xlabel('Year', fontsize=12)
ax.set_title('CONUS Northwestern Forested Mountains Annual Daytime FRP ($W/M^2$), confidence > {}'.format(conf))
labels = [item.get_text() for item in ax.get_xticklabels()]
labels = [str(v) for v in year]
ax.set_xticklabels(labels, rotation=45)
plt.show()

## night
percentiles = np.array([df.night_frp_5.values, df.night_frp_25.values, df.night_frp_median.values, df.night_frp_75.values, df.night_frp_95.values]).T
fig, ax = plt.subplots(figsize=(15,10))
b = customized_box_plot(percentiles, ax, redraw=True, notch=0, sym='+', vert=1, whis=1.5)
ax.set_ylabel('FRP ($W/m^2$)', fontsize=12)
ax.set_xlabel('Year', fontsize=12)
ax.set_title('CONUS Northwestern Forested Mountains Annual Nighttime FRP ($W/M^2$), confidence > {}'.format(conf)
labels = [item.get_text() for item in ax.get_xticklabels()]
labels = [str(v) for v in year]
ax.set_xticklabels(labels, rotation=45)

plt.show()
```

```python

confs=[0, 25, 50, 75]
for c in confs:
    
    frps = []
    for dbf in dbf_files:
        frps.append(extract_frp_day_night_annual(dbf, conf=c)) #default conf=50

    df = pd.DataFrame(frps).sort_values('years')

    ## day
    # percentiles = np.array([frp_day_5, frp_day_25, frp_day_median, frp_day_75, frp_day_95]).T
    percentiles = np.array([df.day_frp_5.values, df.day_frp_25.values, df.day_frp_median.values, df.day_frp_75.values, df.day_frp_95.values]).T
    print(percentiles.shape)
    year = df.years.values

    fig, ax = plt.subplots(figsize=(15,10))
    b = customized_box_plot(percentiles, ax, redraw=True, notch=0, sym='+', vert=1, whis=1.5)
    ax.set_ylabel('FRP ($W/m^2$)', fontsize=12)
    ax.set_xlabel('Year', fontsize=12)
    ax.set_title('CONUS Northwestern Forested Mountains Annual Daytime FRP ($W/M^2$), confidence > {}'.format(c))
    labels = [item.get_text() for item in ax.get_xticklabels()]
    labels = [str(v) for v in year]
    ax.set_xticklabels(labels, rotation=45)

    figname = '../figs/Daytime_FRP_L1_Northwestern_Forested_Mountains_conf_gt_{}'.format(c)
    plt.savefig(figname)
    plt.show()

    ## night
    percentiles = np.array([df.night_frp_5.values, df.night_frp_25.values, df.night_frp_median.values, df.night_frp_75.values, df.night_frp_95.values]).T
    fig, ax = plt.subplots(figsize=(15,10))
    b = customized_box_plot(percentiles, ax, redraw=True, notch=0, sym='+', vert=1, whis=1.5)
    ax.set_ylabel('FRP ($W/m^2$)', fontsize=12)
    ax.set_xlabel('Year', fontsize=12)
    ax.set_title('CONUS Northwestern Forested Mountains Annual Nighttime FRP ($W/M^2$), confidence > {}'.format(c))
    labels = [item.get_text() for item in ax.get_xticklabels()]
    labels = [str(v) for v in year]
    ax.set_xticklabels(labels, rotation=45)

    figname = '../figs/Nighttime_FRP_L1_Northwestern_Forested_Mountains_conf_gt_{}'.format(c)
    plt.savefig(figname)
    plt.show()
```
