```python
import geopandas as gpd
import pandas as pd
import numpy as np
import os, sys
from matplotlib import pyplot as plt
from glob import glob

from helpers import extract_day_night_counts, extract_frp_day_night_annual
from functools import partial
import multiprocessing as mp
import time
```

```python
fire_pts_folders = !ls -d ../modis_fire_points/*
shp_files = [glob(folder + '/*.shp')[0] for folder in fire_pts_folders]
dbf_files = [glob(folder + '/*.dbf')[0] for folder in fire_pts_folders]
```

```python
t0 = time.time()
num_cpu = mp.cpu_count() - 2
num_cpu = 3
pool = mp.Pool(processes=num_cpu)

# extract day night counts
#a = pool.map(partial(extract_day_night_counts, conf=50), shp_files)

# extract FRP values for day night
a = pool.map(partial(extract_frp_day_night_annual, conf=50), dbf_files)


pool.close()
pool.join()
t1 = time.time()

print_time = (t1-t0)/60
print('time: {} minutes'.format(print_time))
```

```python
import pickle
with open('mp_result_frp_annual_day_night.pickle', 'wb') as f:
    pickle.dump(a, f)
```

## process result

```python
year = [int(i[0][1:]) for i in a]
years_sorted = np.argsort(year)
year = [year[i] for i in years_sorted]

frp_day_mean = [i[1] for i in a]
frp_day_mean = [frp_day_mean[i] for i in years_sorted]

frp_day_std = [i[2] for i in a]
frp_day_std = [frp_day_std[i] for i in years_sorted]

frp_night_mean = [i[3] for i in a]
frp_night_mean = [frp_night_mean[i] for i in years_sorted]

frp_night_std = [i[4] for i in a]
frp_night_std = [frp_night_std[i] for i in years_sorted]
```

```python
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(15,10))
plt.errorbar(year, frp_night_mean, frp_night_std, fmt='ok', capsize=3, lw=3, label='nighttime')
plt.errorbar(year, frp_day_mean, frp_day_std, fmt='or', capsize=3, lw=3, label='daytime')
plt.xticks(year, fontsize=14)
plt.legend()
```

```python
frp_df = pd.DataFrame({'year':sorted(year),
                      'day_mean': frp_day_mean,
                      'day_std': frp_day_std,
                      'night_mean': frp_night_mean,
                      'night_std': frp_night_std})
```

```python
frp_df.day_mean.plot.bar(x=frp_df.year, yerr=frp_df.day_std, capsize=4)

#means.plot.bar(yerr=errors, ax=ax, capsize=4)
```

```python
ax = frp_df.plot(kind='bar', x='year', y='day_mean', yerr='day_std', capsize=5, color='r')
ax = frp_df.plot(kind='bar', x='year', y='night_mean', yerr='night_std', capsize=5, ax=ax)
```

```python
means = frp_df[['year', 'day_mean', 'night_mean']]
means.set_index('year', inplace=True)

errors = frp_df[['year', 'day_std', 'night_std']]
errors.set_index('year', inplace=True)
errors = errors.rename(columns={'day_std':'day_mean', 
                        'night_std':'night_mean'})

#plot
fig, ax = plt.subplots()
means.plot.bar(yerr=errors, 
               capsize=4, figsize=(15,10), rot=45, fontsize=14, alpha=0.25, edgecolor='k', ax=ax)
plt.legend(fontsize=12)
```

```python
means.plot(kind='line', style='--o', figsize=(15,10), rot=45, fontsize=12)
plt.ylabel('Mean FRP', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.xticks(year);
```
