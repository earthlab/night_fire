```python
import rasterio as rio
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
from earthpy.clip import clip_shp
```

```python
import os,sys
from glob import glob
```

```python
geomac_folder = '../shp/carr_geomac/'
zips = glob(geomac_folder + '/*.zip')
df_zips = ['zip://' + f for f in zips]
```

```python
days = [os.path.basename(f).split('_')[2:4] for f in zips]
```

```python
days_arr = np.array(days)
```

```python
import pandas as pd
```

```python
days_df = pd.DataFrame({'ymd': days_arr[:,0], 'time': days_arr[:,1]})
days_df_grp = days_df.groupby('ymd')
```

```python
# read the 2018 MODIS AF file
modis_af_2018 = '../modis_fire_points/DL_FIRE_M6_50657/fire_archive_M6_50657.shp'
df_modis = gpd.read_file(modis_af_2018)
df_modis['month'] = [int(acq.split('-')[1]) for acq in df_modis['ACQ_DATE']]
df_modis['day'] = [int(acq.split('-')[2]) for acq in df_modis['ACQ_DATE']]
```

```python
# for ind, fi in enumerate(df_zips):
for nm,grp in days_df_grp:
    ind = grp.index[-1]
    bfile = df_zips[ind]
    df_geomac = gpd.read_file(bfile)


    # clipped_df = clip_shp(df_modis, df_geomac.unary_union)
    # print(clipped_df.shape, df_modis.shape)

    # subset by DOY first
    ymd = days[ind][0]
    geomac_day = int(ymd[-2:])
    geomac_month = int(ymd[4:6])
    
    # plot the perimeter
    df_geomac.plot()
    plt.show()
    
    # subset the data frame
    df_sub = df_modis.query('month == {}'.format(geomac_month))
    print(geomac_month, df_sub.shape, df_modis.shape)

    df_sub = df_modis.query('month == {} & day == {}'.format(geomac_month, geomac_day))
    print(geomac_month, geomac_day, df_sub.shape, df_modis.shape)
    
    # clip the data frame
    geomac_points = clip_shp(df_sub, df_geomac)
    
    # plot the CONFIDENCE values for the clipped and subsetted DF by DAYNIGHT
    dn_geomac = geomac_points.groupby('DAYNIGHT')
    plt.figure(figsize=(10,10))
    bins=np.arange(0,101,5)
    for nm, grp in dn_geomac:

        if nm == 'D':
            plt.hist(grp['CONFIDENCE'], bins=bins, alpha=0.4, edgecolor='k', label=nm, color='r')
        else:
            plt.hist(grp['CONFIDENCE'], bins=bins, alpha=0.4, edgecolor='k', label=nm, color='b')

    plt.legend()
    plt.grid(True)
    plt.show()
```
